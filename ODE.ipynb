#<h1><center><b>RUNGE-KUTTA METHOD</b>

---

## RK-2
Runge-Kutta provides you highly meticluously woven expressions to solve LDEs  
<center>$y(t+\triangle t) = y(t) + \triangle t \; .\, y'(y(t) + \frac{k(t)}{2}, t+\frac{\triangle t}{2}) + \bigcirc (\triangle t^3)$

where

$k(t)=\triangle t . \, y'(y,t)$
</center> 

other symbols have usual meaning. This expression (called, RK-2) says that value of function at some interval later is sum of function itself and increment when the slope is calculated at the mid-point of the interval.

The problem with euler's technique is evidently the order of approximation it undertakes.
So Range-kutta used higher order derivative to calculate function after successive intervals. 

## ***Origin of expression***

From taylor's expansion

$y(t + \triangle t) y(t) + \triangle t\, .\, y'(y,t) + \frac{1}{2}\triangle t^2 y''(y', \,y,\, t) + \bigcirc (\triangle t^3), \tag{1}$

We will omit error bar hitherto
<br>

<center>$y(t + \triangle t) = y(t) + \triangle t\, .[\, y'(t) + \frac{1}{2}\triangle t \, y''] \tag{2} $</center>

The term $ (y'(t) + \frac{1}{2}\triangle t \: y''(y,t))$ can be thought as the slope of the function at half interval later.   *(Note now the function is taking more than one arguments)*

i.e.

Let,  $g(t) = y'(t) $

$\implies g(t+ \frac{1}{2} \triangle t) = g(t) + \frac{1}{2} \triangle t \, g'(t)  $
which is exactly the expression as above. Therefore, the second equation can be written as 

<center>$y(t + \triangle t) = y(t) + \triangle t\, .y'(y + \frac{1}{2}\triangle t. y',\:t +  \frac{1}{2}\triangle t)  \tag{3} $</center>

The equation (3) is often called RK-2, as we have another Range-kutte formula which uses 5th order of approximation and so is called RK-4. ***RK-4*** uses four intermediate slopes to carry out the increment in the function.<br><br>

## RK-4

\begin{equation}
y(t+\triangle t) = y(t) + \frac{1}{6} [k_1+2k_2+2k_3+k_4] + \bigcirc (\triangle t^5) \\
where \\ 
\\
k_1 = \triangle t.\;y'(y,t) \\
k_2 = \triangle t.\;y'(y + \frac{1}{2} k_1,t + \frac{1}{2} \triangle t) \\
k_3 = \triangle t.\;y'(y + \frac{1}{2} k_2,\:t + \frac{1}{2} \triangle t)) \\
k_4 = \triangle t.\;y'(y +  k_3, \; t + \triangle t) 
\end{equation}

# <center><b><h1>VERLET'S METHOD </b></center>

---

Now we will be talking about methods which are mystifying in the way that they give suprising accuracies in certain cases. We start by describing the Verlet's algorithm. 

Let a particle in a system having parameters given, i.e. its velocity $v$ and position $p$ is given at some time $t_0$. The particle is in the conservative field, therefore acceleration of the particle is only dependent on position of the particle ($a = a(p))$.  Then we can use following method to find its further configuration:
\begin{align}
p(t + \frac{1}{2} \triangle t) &= p(t) + \frac{1}{2} \triangle t. \; v(t) \tag{1}, \\
v(t + \triangle t) &= v(t) + \triangle t. \; a(\,p( t)\,) \tag{2}  \\
p(t + \triangle t) &= p(t + \frac{1}{2} \triangle t) + \frac{1}{2} \triangle t. \; v(t + \triangle t) \tag{3}\\
\end{align}

Specifically, this set of equations is called position-verlet. Symmetrically you can also obtain velocity-verlet. This method gives no error in *energy* values, even when it is using the third order approximation. It also has memory efficiency over other algorithm. The only disadvantage is limited applicabality.

#TIPS FOR WRITING THE CODE

We will show how to write the problems while providing you a model problem of *Forced damped harmonic oscillator*.

 For instance,

<tt> Consider a bob of mass '$m$' is weld to a very light rod of lenght $l$, which in turn is fitted to a wall such that it can pivot about one end (opposite end to the bob). The whole set-up is submerged in a viscous liquid on earth such that it provides the drag to the object which is proportional to the object's velocity ($b$ being the proportionality constant). At the pivot there is some type of motor which is making the rod to go on oscillate. We can always align our coordinate system such that the motion happens in x-y plane. Setting up the suitable coordinate system and applying Newton's laws, we get the differential equation for this situation as following:
</tt>

\begin{equation}
\frac{d^2{\theta}}{dt^2} -{b}\frac{d{\theta}}{dt}-\omega_0^2sin{\theta}=\frac{F(t) }{m}
\end{equation}

> where

>>>>    $w_0= \sqrt{\frac{g}{l}}$  .... called angular frequency

>>>>   $ T_0 = \frac{1}{\omega_0} $ .... called Time period

>>>>   $b$ ..... called the damping constant for unit mass

>>>>   $F(t)$  ..... is external angular force which depends on time.


before going directly to code we would like to non-dimensioanlise the equation. We'll assume that the motor's mechanism is exerting an acceleration equal to $F_0cos\lambda t$ where $\lambda$ could be the natural frequency of the motor. 
<br>
remark : we are using acceleration in loose terms, you can't really accelerate something directly. It is force only which make interactions betweem systems.  

\begin{equation}
T = t/T_0 = t\omega_0  \\
k = b/\omega_0 = b\,T_0   \\
X = \theta     \\
\Lambda = \frac{\lambda}{\omega_0} \\
\tilde{f} = F_0/\omega_0^2
\end{equation}


which gives your equation as 

\begin{equation}
  \frac{d^2X}{dT^2} + k \frac{dX}{dT} +sinX = \tilde{f}cos(\Lambda T)
\end{equation}

This is the final differential equation we have in hand and we are going to solve it with the above described algorithms now. 

```
#####importing libraries 
## we will be needing two libraries one being about plotting things out 
##(as their name suggests) and other is for using predefined mathematical functions. 

from matplotlib import pylab as plt   
import math as m

###### INITIAL CONDITIONS and SYSTEM PARAMETERS
## we have already discussed how numerical solutions build-upon initial conditions
## and thus providing them here. 

# Ratio of damping to natural frequency
k=  0.1
# Ratio of system's inherent time scale to external force's time scale
lam = .8
# non-dimensional external acceleration 
f = .10


## the next step is particularly used in numerical solutions 'cause you will need 
## solution till some point, so that we can solve it in for finite time. So our simulation
## will run till T_end (it's T_end times the time period ptherwise)
T_end = 100    
## the step of time would del_T (unit increment in time)
del_T = .01
## therefore the number of iteration you need to reach T_end
N = int( T_end / del_T )
```

```
#### DEFINING VARIABLE and their 'HISTORIANS'

##Defining velocity as a function
def Vel(X, Y, T):
    return Y

##Defining acceleration as a function
def acc(X, Y, T):
    return f*m.cos(lam*T)-m.sin(X) -(k * Y) 
```

Now we will try this out with different algorithms and will be able to compare them. 

```
#### SOLVING DE with euler 

# initial position
X = 1.0
# initial velocity
Y = 0.0
# # initial time
T = 0.0

## Now since variables will be changing their values with each iteration 
## we are defining some lists (arrays) so that we can plot the trajectory, saving 
## values of each variable after every time step that's why calling them historians 

T_list = []
X_list = []
Y_list = []


for i in range (N):
## if looked at it carefully, you can see two options to increment first; position and velocity.
## which comes due to the coupling of two DEs.
## We will implement them simultaneosly by using temperory variable. See carefully how this is simulataneous.. 
    
    X_temp = X + del_T*Y
    Y=Y +del_T*acc(X,Y,T)     
    X=X_temp
    T +=del_T

    X_list.append(X)
    Y_list.append(Y)
    T_list.append(T)

# you have the time evolution of motion (position and velocity)

# plotting 
plt.title("Euler's Algorithm")

# this is main command for plotting anything else is decoration
plt.plot(T_list, X_list, color='orange',label='position')
plt.plot(T_list, Y_list, color='green',label='velocity')
plt.legend()
plt.xlabel('time')
# this is to show your graphs. It will show on a single graph all the plots you have plotted till now.
plt.show()
```

```
#### SOLVING DE with RK-2

# Doing this step all over again because we don't want the final conditions of previous algorithm to be initial of this.

# initial position
X = 1.0
# initial velocity
Y = 0.0
# # initial time
T = 0.0

T_list = []
X_list = []
Y_list = []

# as we have discussed previously how to tackle with coupled differential equation 
# we created two increments k1 and k2 (we will call 'k's as increment even though they can represent decrement but essentially it's encompassing change)

for i in range(N):
    k1=del_T*Vel(X,Y,T)
    k2=del_T*acc(X,Y,T)
    X=X+del_T*Vel(X+k1/2, Y+k2/2, T+del_T/2)
    Y=Y+del_T*acc(X+k1/2, Y+k2/2, T+del_T/2)
    T=T+del_T

    X_list.append(X)
    Y_list.append(Y)
    T_list.append(T)

# going with the same thing again
plt.title("RK-2 Algorithm")
plt.xlabel("time")
plt.plot(T_list,X_list,color='red',label='angular displacement')
plt.plot(T_list,Y_list,color='blue',label='angular velocity')
plt.legend()
plt.show()
```

```
#### SOLVING DE with RK-4

# initial position
X = 1.0
# initial velocity
Y = 0.0
# # initial time
T = 0.0

T_list = []
X_list = []
Y_list = []


for i in range(N):

    K1 = del_T * acc(X, Y, T)
    K5 = del_T * Vel(X, Y, T)
    K2 = del_T * acc(X + Y * del_T / 2, Y + K1 / 2, T + del_T / 2)
    K6 = del_T * Vel(X + Y * del_T / 2, Y + K5 / 2, T + del_T / 2)
    K3 = del_T * acc(X + Y * del_T / 2, Y + K2 / 2, T + del_T / 2)
    K7 = del_T * Vel(X + Y * del_T / 2, Y + K6 / 2, T + del_T / 2)
    K4 = del_T * acc(X + Y * del_T / 2, Y + K3 / 2, T + del_T / 2)
    K8 = del_T * Vel(X + Y * del_T / 2, Y + K7 / 2, T + del_T / 2)

    # we have calculated the increments and according to rule we are going to update our variables
    X = X + (1 / 6) * (K5 + 2 * K6 + 2 * K7 + K8)
    Y = Y + (1 / 6) * (K1 + 2 * K2 + 2 * K3 + K4)
    T = T + del_T

    X_list.append(X)
    Y_list.append(Y)
    T_list.append(T)

plt.title("RK-4 Algorithm")
plt.xlabel("time")
plt.plot(T_list,X_list,color='gray',label='angular displacement')
plt.plot(T_list,Y_list,color='purple',label='angular velocity')
plt.legend()
plt.show()
```

 We can do all sort of things with these results, not limiting ourselves to only plot time evolution. Later we will discuss chaos theory and these are going to be pretty helpful there. Also the error analysis of the results. For our solution to be more physical we put another constraint on the process that of energy conservation. Newton's law have in-built consistency with conservation laws but our simulation doesn't. We'll move ahead with verlet algorithm and use a different situation since it is not a good choice for non-conservative forces as they don't depend on position as you would remember. 
We are going to analyse the energy of the pendulum in this problem to demonstrate the brilliance of verlet in energy conservation. 

```
# verlet algorithm

# imagine a simpler situation when you have simply an oscillating pendulum in vacuum and fixed pivot. 
# defining the acceleration
def A(X):
    return -4*m.pi*m.pi*m.sin(X)  #acceleration

# defining the energy
# this is essentially total energy, it is taking two variable because to define each state you always need both position and velocity of particle.
def erg(x, v):
    return (0.5 * v * v + 0.5 * 4 * m.pi * m.pi * x ** 2) / (2 * m.pi * m.pi)

# INITIAL CONDITIONS and PARAMETERS of SIMULATION
X=X_=1.0
V=V_=0.0
T=0.0
end_T=10
d_T=0.01
N=int(end_T/d_T)


# again defining our historians 
T_list = []
X_list = []
V_list = []
X__list = []
V__list = []
E_list = []
E__list = []


# main loop where we are implimenting our algorithm recursively.
# you can omit 0 in the paranthesis of loop since it's default initial value is 0.
for i in range (0,N):  
    # position verlet
    V_ = V_ + d_T * 0.5 * A(X_)
    X_ = X_ + d_T * V_
    V_ = V_ + d_T * 0.5 * A(X_)
    # velocity verlet
    X = X + d_T * 0.5 * V
    V = V + d_T * A(X)
    X = X + d_T * 0.5 * V


    T=T+d_T

    X_list.append(X)
    V_list.append(V)
    E_list.append(erg(X,V))

    T_list.append(T)

    X__list.append(X_)
    V__list.append(V_)
    E__list.append(erg(X_,V_))


plt.title("comparison of velocity and position Algorithm")
plt.xlabel("time")
plt.plot(T_list,X_list,color='orange',label='angular displacement from position verlet')
plt.plot(T_list,X__list,color='purple',label='angular displacement from velocity verlet')
plt.legend()
plt.show()

## total Energy 
# you can plot simply kinetic energy but essentially we want to see it's error. 
# also we can use any algorithm it doesn't matter in energy analysis
plt.title("Total energy v/s Time")
plt.xlabel("time")

# you should be thinking we didn't add it before, you can see why we added it now by removing it.
plt.ylim(0,2)
plt.plot(T_list,E_list,color='black',label='total energy from position verlet')
plt.legend()
plt.show()
```
